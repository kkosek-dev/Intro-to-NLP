```python
import selenium
import splinter
from splinter import Browser
from bs4 import BeautifulSoup
from pprint import pprint
import json
```

```python
browser = Browser('chrome')
```

```python
browser.visit('https://www.walmart.com/reviews/product/108684613')
```

```python
html = browser.html
```

```python
soup = BeautifulSoup(html,"html.parser")
```

```python
review_list = []
rows = soup.find_all("li", class_="dib w-100 mb3")
soup.find('span', class_='w_iUH7')
```

```python
review_list = []
rows = soup.find_all("li", class_="dib w-100 mb3")

for i in rows: 
    date = i.find('div', class_='f7 gray mt1').text
    text = i.find('span', class_='tl-m mb3 db-m').text
    stars = i.find('span', class_='w_iUH7').text
    review_list.append( {
        'date':date,
        'text':text,
        'stars':stars
    } )
    
```

```python
import pandas as pd
pd.DataFrame(review_list)
```
